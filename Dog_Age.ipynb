## Dog Age vs. Human Age

```python
import numpy as np 
import matplotlib.pylab as plt
%matplotlib inline 
```

```python
i = float(input('Enter the age of your dog in years:'))
dog_age = ((0.010255*i**3) - (0.340119*i**2) + (7.397499*i) + (6.618361))
print('The equivalent age of your dog in human years is:', format(dog_age, '.1f'))
```

```python
print("Mappings from chronological years to dog age:")
for i in range(1,18):
    dog_age = ((0.010255*i**3) - (0.340119*i**2) + (7.397499*i) + (6.618361))
    print(i,":", format(dog_age, '.1f'), end="; ")
```

```python
dog_result = np.empty((0, 25))
for i in range(0,25):
    dog_age = ((0.010255*i**3) - (0.340119*i**2) + (7.397499*i) + (6.618361))
    dog_result = np.append(dog_result, [dog_age])
```

```python
dog_result
```

```python
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(dog_result, color="steelblue", label="Dog Age")
ax.set_title("Estimate of Dog Age vs. Human in Years", fontsize=20)
ax.set_xlabel("Years", fontsize=16)
ax.set_ylabel("Age", fontsize=16)
ax.legend(loc='upper left')

ax.grid(True, alpha=0.25)
```
